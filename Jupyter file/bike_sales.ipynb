```python
import pandas as pd
df = pd.read_csv("uncleaned bike sales data(Bike Sales).csv")
df.head()
```

Loads the pandas library as well as the pandas file.
Shows the first 5 rows


```python
df.info()
```

Shows which columns have missing values (important).

Shows column types (int, float, object).

```python
df.describe(include='all')
```

.isnull().sum() → exactly how many missing values

.describe() → summary of stats

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.describe(include='all')
```

Convert string → real date
Invalid dates become NaT (missing datetime)

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.replace("-", "_")
df.columns
```

changes the column names from empty spaces to _ and - to _

```python
df.isnull().sum()
```

shows how many nulls values are there in each column

```python
df.duplicated().sum() 
df = df.drop_duplicates()  
df.shape     
df.duplicated().sum()
```

checks the no of duplicate values and drops duplicates


```python
money_cols = ['Unit_Cost', 'Unit_Price', 'Profit', 'Cost', 'Revenue']

for col in money_cols:
    df[col] = df[col].str.replace("$", "").str.replace(",", "")
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df[money_cols].head()
df.info()    # these columns should now be float64
```

'Unit_Cost', 'Unit_Price', 'Profit', 'Cost', 'Revenue' we changed these columns data type values from string to float

```python
df['Age_Group'].value_counts()   # check categories first

df['Age_Group'] = df['Age_Group'].fillna("Unknown")
 
```

```python
df['Age_Group'].isnull().sum()
df['Age_Group'].value_counts()
```

we changed the missing Nan value from age group column to 'unknown'

```python
median_day = df['Day'].median()
df['Day'] = df['Day'].fillna(median_day)
```

```python
df['Day'].isnull().sum()
df['Day'].head()
```

there is a missing value in day column so we fill it with a median value of the day column

```python
df['Sub_Category'] = df['Sub_Category'].fillna("Unknown")
df['Product_Description'] = df['Product_Description'].fillna("No Description")
```

```python
df.rename(columns={'Sales_Order_#': 'Sales_Order'}, inplace=True)
```

sub category and product description column also has nan values so we fill it with values unknown and no decription respectively

```python
cat_cols = ['Customer_Gender','Country','State','Product_Category','Sub_Category']

for col in cat_cols:
    df[col] = df[col].str.strip().str.title()

```

```python
df[cat_cols].head()
```

removes spaces and unifies case etc

```python
df.to_csv("bike_sales_cleaned.csv", index=False)
```

saves this data to a new csv file 

```python
df.info()
df.isnull().sum()
df.head()
df.describe(include='all')
```

checks if datatypes are correct

missing values handled

numbers converted

no duplicates

```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("bike_sales_cleaned.csv")
df.head()
```

```python
df.groupby('Date')['Revenue'].sum().plot(kind='line', figsize=(10,5))
plt.title('Revenue Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.tight_layout()
plt.savefig('revenue_trend.png')   # saves in your working folder
plt.show()
```

LINE GRAPH/LINE PLOT GRAPH

```python
df.groupby('Product_Category')['Revenue'].sum().sort_values().plot(kind='bar', figsize=(10,5))
plt.title('Total Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.tight_layout()
plt.savefig('revenue_by_category.png')
plt.show()
```

BAR GRAPH

```python
df.plot(kind='scatter', x='Unit_Price', y='Profit', alpha=0.6, figsize=(8,5))
plt.title('Unit Price vs Profit')
plt.xlabel('Unit Price')
plt.ylabel('Profit')
plt.tight_layout()
plt.savefig('unitprice_vs_profit.png')
plt.show()
```

SCATTER PLOT


```python
df.boxplot(column='Revenue', by='Product_Category', figsize=(10,6), rot=45)
plt.title('Revenue Distribution by Product Category')
plt.suptitle('')
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.tight_layout()
plt.savefig('revenue_boxplot.png')
plt.show()
```

BOX PLOT


```python
df['Customer_Age'].plot(kind='hist', bins=10, figsize=(8,5))
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('age_histogram.png')
plt.show()
```
